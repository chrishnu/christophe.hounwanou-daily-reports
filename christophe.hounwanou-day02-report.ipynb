### Exercise 33:
Write a Python function that reads a file and prints the longest line in the file.

```python

def p_longest_line(filename):
    with open(filename, 'r') as file:
        lines = file.readlines()
        longest_line = max(lines, key=len)
        print(f"Longest line: {longest_line.strip()}")

p_longest_line("filename.txt")

```

### Exercise 34:
Write a Python script to count the frequency of each letter in a text file `"message.txt"`.

```python

from collections import Counter

with open("message.txt", 'r') as file:
    content = file.read().lower()
    letters = [char for char in content if char.isalpha()]
    frequency = Counter(letters)
    print(frequency)

```

### Exercise 37:
Write a Python function that reads a file and removes all punctuation marks from the text.

```python
import string

def remove_punctuation(filename):
    with open(filename, 'r') as file:
        content = file.read()
        return content.translate(str.maketrans('', '', string.punctuation))

print(remove_punctuation("filename.txt"))

```

### Exercise 38:
Create a Python function that writes a list of strings to a file, each string on a new line.

### Exercise 43:
Write a Python script to find the most frequent word in a text file `"book.txt"`.

```python

from collections import Counter

with open("book.txt", 'r') as file:
    words = file.read().split()
word_count = Counter(words)

most_common_word = word_count.most_common(1)[0]

print(f"Most frequent word: {most_common_word[0]} with {most_common_word[1]} occurrences.")

```

### Exercise 47:
Create a Python function that opens a file and returns the number of unique words in the file.

```python

def unique_word_count(filename):
    with open(filename, 'r') as file:
        words = file.read().split()
    return len(set(words))

print("Number of unique words:", unique_word_count("file.txt"))

```

### Exercise 49:
Write a Python function that reads a file and returns the most common letter in the text.

```python
# Exercise 49

from collections import Counter
import string

def most_common_letter(filename):
    with open(filename, 'r') as file:
        content = file.read().lower()
    letters = [char for char in content if char in string.ascii_lowercase]
    letter_count = Counter(letters)
    return letter_count.most_common(1)[0]

print("Most common letter:", most_common_letter("file.txt")[0])

```
